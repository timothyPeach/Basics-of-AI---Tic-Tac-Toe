```python
import random as rn
#import time

#6 rows 7 columns
board = ["no"]+["_"]*42 +["no"]

print("Flipping coin to see who goes first!")

gameOver = False

goFirst = True
coin = rn.randint(0,1)
if coin == 0:
    goFirst = False
    
if goFirst == True:
    print("You go first!")
else:
    print("The computer goes first!")
    
def printBoard():
    print("===")
    for i in range(0,6):
        print("["+board[(7*i)+1]+","+board[(7*i)+2]+","+board[(7*i)+3]+","+board[(7*i)+4]+","+board[(7*i)+5]+","+board[(7*i)+6]+","+board[(7*i)+7]+"]")
    print("===")
    
def playerMove():
    try:
        move = int(raw_input("Enter an integer 1 thru 42 to determine your move."))
        if board[move] == "_":
            board[move] = "X"
        elif board[move] not in range(1,43):
            print("Hey stick to numbers on the board, lol!")
            playerMove()
        else:
            print("There's already a move there!")
            playerMove()
    except:
        print("You're trying to enter in something crazy, stop dat lol!")
        playerMove()

def computerMove():
    legalMoves = []
    for i in range(0,7):
        if board[36+i] == "_":
            legalMoves.append(36+i)
        elif board[29+i] == "_":
            legalMoves.append(29+i)
        elif board[22+i] == "_":
            legalMoves.append(22+i)
        elif board[15+i] == "_":
            legalMoves.append(15+i)
        elif board[8+i] == "_":
            legalMoves.append(8+i)
        elif board[1+i] == "_":
            legalMoves.append(1+i)
    move = rn.choice(legalMoves)
    if board[move] == "_":
        board[move] = "0"
    elif board[move] not in range(1,43):
        print("The computer is trying to make an illegal move! Weird!")
        computerMove()
    else:
        print("The computer is trying to go where a move already exists! Lol!")
        computerMove()
        
def scoreBoard(aBoard):
    lanes = [[1,2,3,4],[2,3,4,5],[3,4,5,6],[4,5,6,7],
             [8,9,10,11],[9,10,11,12],[10,11,12,13],[11,12,13,14],
             [15,16,17,18],[16,17,18,19],[17,18,19,20],[18,19,20,21],
             [22,23,24,25],[23,24,25,26],[24,25,26,27],[25,26,27,28],
             [29,30,31,32],[30,31,32,33],[31,32,33,34],[32,33,34,35],
             [36,37,38,39],[37,38,39,40],[38,39,40,41],[39,40,41,42],
             #/horz
             [1,8,15,22],[8,15,22,29],[15,22,29,36],
             [2,9,16,23],[9,16,23,30],[16,23,30,37],
             [3,10,17,24],[10,17,24,31],[17,24,31,38],
             [4,11,18,25],[11,18,25,32],[18,25,32,39],
             [5,12,19,26],[12,19,26,33],[19,26,33,40],
             [6,13,20,27],[13,20,27,34],[20,27,34,41],
             [7,14,21,28],[14,21,28,35],[21,28,35,42],
             #/vert
             [15,23,31,39],[8,16,24,32],[16,24,32,40],
             [1,9,17,25],[9,17,25,33],[17,25,33,41],
             [2,10,18,26],[10,18,26,34],[18,26,34,42],
             [3,11,19,27],[11,19,27,35],[4,12,20,28],
             #/bs diag
             [4,10,16,22],[5,11,17,23],[11,17,23,29],
             [6,12,18,24],[12,18,24,30],[18,24,30,36],
             [7,13,19,25],[13,19,25,31],[19,25,31,37],
             [14,20,26,32],[20,26,32,38],[21,27,33,39]]
             
    laneStatus = [""]*69

    for i in range(len(lanes)):
        xCount = 0
        zeroCount = 0
        for j in lanes[i]:
            if aBoard[j] == "X":
                xCount += 1
            if aBoard[j] == "0":
                zeroCount += 1
        if xCount == 2 and zeroCount == 0:
            laneStatus[i] = "Danger"
        if xCount == 0 and zeroCount == 2:
            laneStatus[i] = "Great"
        if xCount == 0 and zeroCount == 3:
            for q in range(len(lanes[i])):
                if lanes[i][q] == "_" and lanes[i][q] > 35:
                    laneStatus[i] = "Amazing+"
                if lanes[i][q] == "_" and lanes[i][q] < 36 and lanes[i+7][q] != "_":
                    laneStatus[i] = "Amazing+"
        if xCount == 3 and zeroCount == 0:
            for q in range(len(lanes[i])):
                if lanes[i][q] == "_" and lanes[i][q] > 35:
                    laneStatus[i] = "Peril+"
                if lanes[i][q] == "_" and lanes[i][q] < 36 and lanes[i+7][q] != "_":
                    laneStatus[i] = "Peril+"
        if xCount == 0 and zeroCount == 3 and laneStatus[i] != "Amazing+":
            laneStatus[i] = "Amazing-"
        if xCount == 3 and zeroCount == 0 and laneStatus[i] != "Peril+":
            laneStatus[i] = "Peril-"
        if xCount == 0 and zeroCount == 4:
            laneStatus[i] = "Victory"
        if xCount == 4 and zeroCount == 0:
            laneStatus[i] = "Defeat"
        if xCount == 0 and zeroCount == 0:
            laneStatus[i] = "Empty"
        if xCount == 1 and zeroCount == 0:
            laneStatus[i] = "Theirs"
        if xCount == 0 and zeroCount == 1:
            laneStatus[i] = "Ours"
        if xCount > 0 and zeroCount > 0:
            laneStatus[i] = "Dead"

        #for i in laneStatus:
            #print(i)

        # So this process should yield a score.

    currentScore = 0

    for i in laneStatus:
        if i == "Theirs":
            currentScore -= 1
        if i == "Ours":
            currentScore += 1
        if i == "Danger":
            currentScore -= 100
        if i == "Great":
            currentScore += 100
        if i == "Amazing-":
            currentScore += 10000
        if i == "Peril-":
            currentScore -= 10000
        if i == "Amazing+":
            currentScore += 1000000
        if i == "Peril+":
            currentScore -= 1000000
        if i == "Victory":
            currentScore += 100000000
        if i == "Defeat":
            currentScore -= 100000000

    #print("currentScore: " + str(currentScore))
    return currentScore

def returnLegalMoves(aBoard):
    legalMoves = []
    for i in range(0,7):
        if aBoard[36+i] == "_":
            legalMoves.append(36+i)
        elif aBoard[29+i] == "_":
            legalMoves.append(29+i)
        elif aBoard[22+i] == "_":
            legalMoves.append(22+i)
        elif aBoard[15+i] == "_":
            legalMoves.append(15+i)
        elif aBoard[8+i] == "_":
            legalMoves.append(8+i)
        elif aBoard[1+i] == "_":
            legalMoves.append(1+i)
    return legalMoves

def strongMove():
    legalMoves = returnLegalMoves(board)
            
    #print(legalMoves)
    
    move = 0
    
    lanes = [[1,2,3,4],[2,3,4,5],[3,4,5,6],[4,5,6,7],
             [8,9,10,11],[9,10,11,12],[10,11,12,13],[11,12,13,14],
             [15,16,17,18],[16,17,18,19],[17,18,19,20],[18,19,20,21],
             [22,23,24,25],[23,24,25,26],[24,25,26,27],[25,26,27,28],
             [29,30,31,32],[30,31,32,33],[31,32,33,34],[32,33,34,35],
             [36,37,38,39],[37,38,39,40],[38,39,40,41],[39,40,41,42],
             #/horz
             [1,8,15,22],[8,15,22,29],[15,22,29,36],
             [2,9,16,23],[9,16,23,30],[16,23,30,37],
             [3,10,17,24],[10,17,24,31],[17,24,31,38],
             [4,11,18,25],[11,18,25,32],[18,25,32,39],
             [5,12,19,26],[12,19,26,33],[19,26,33,40],
             [6,13,20,27],[13,20,27,34],[20,27,34,41],
             [7,14,21,28],[14,21,28,35],[21,28,35,42],
             #/vert
             [15,23,31,39],[8,16,24,32],[16,24,32,40],
             [1,9,17,25],[9,17,25,33],[17,25,33,41],
             [2,10,18,26],[10,18,26,34],[18,26,34,42],
             [3,11,19,27],[11,19,27,35],[4,12,20,28],
             #/bs diag
             [4,10,16,22],[5,11,17,23],[11,17,23,29],
             [6,12,18,24],[12,18,24,30],[18,24,30,36],
             [7,13,19,25],[13,19,25,31],[19,25,31,37],
             [14,20,26,32],[20,26,32,38],[21,27,33,39]]
    
    currentScore = scoreBoard(board)
        
    tempBoards = [0]*len(legalMoves)
    tempScores = [0]*len(legalMoves)    
    
    for i in range(len(tempBoards)):
        tempBoards[i] = []
            
    #print(tempBoards)
    
    for i in range(len(tempBoards)):
        tempBoards[i] = [0]*(len(board)-1)
       
    for x in range(len(tempBoards)):
        for i in range(1,43):
            tempBoards[x][i] = board[i]
            
    for i in range(len(tempBoards)):
        tempBoards[i][legalMoves[i]] = "0"
    
    print(tempBoards)
    
    tempBoards2 = [0]*len(tempBoards)
    
    for i in range(len(tempBoards)):
        tempBoards2[i] = tempBoards[:]
    
    legalMoves2 = []
    
    for i in range(len(tempBoards2)):
        legalMoves2 = returnLegalMoves(tempBoards[i])
        for x in range(len(tempBoards2[i])):
            #print(tempBoards2[i][x])
            tempBoards2[i][x][legalMoves2[x]] = "X"
            
    
    print(tempBoards2)        
    
    #tempBoards3 = [tempBoards2]*len(legalMoves)
    
    #for i in range(len(legalMoves)):
        #tempScores[i] = scoreBoard(tempBoards[i])
        
    #for i in range(len(tempScores)):
        #print("move: " + str(legalMoves[i]) + "   score: " + str(tempScores[i]))
        
    #move = legalMoves[tempScores.index(max(tempScores))]
    
    #and now we move like before:
    
    if board[move] == "_":
        board[move] = "0"
    elif board[move] not in range(1,43):
        print("The computer is trying to make an illegal move! Weird!")
        #but this time if it screws up, we want it to at least be able to move if it messes up.
        computerMove()
    else:
        print("The computer is trying to go where a move already exists! Lol!")
        #that way the player won't have to sit there, idk.
        computerMove()
        
def checkIfGameOver():
    lanes = [[1,2,3,4],[2,3,4,5],[3,4,5,6],[4,5,6,7],
             [8,9,10,11],[9,10,11,12],[10,11,12,13],[11,12,13,14],
             [15,16,17,18],[16,17,18,19],[17,18,19,20],[18,19,20,21],
             [22,23,24,25],[23,24,25,26],[24,25,26,27],[25,26,27,28],
             [29,30,31,32],[30,31,32,33],[31,32,33,34],[32,33,34,35],
             [36,37,38,39],[37,38,39,40],[38,39,40,41],[39,40,41,42],
             #/horz
             [1,8,15,22],[8,15,22,29],[15,22,29,36],
             [2,9,16,23],[9,16,23,30],[16,23,30,37],
             [3,10,17,24],[10,17,24,31],[17,24,31,38],
             [4,11,18,25],[11,18,25,32],[18,25,32,39],
             [5,12,19,26],[12,19,26,33],[19,26,33,40],
             [6,13,20,27],[13,20,27,34],[20,27,34,41],
             [7,14,21,28],[14,21,28,35],[21,28,35,42],
             #/vert
             [15,23,31,39],[8,16,24,32],[16,24,32,40],
             [1,9,17,25],[9,17,25,33],[17,25,33,41],
             [2,10,18,26],[10,18,26,34],[18,26,34,42],
             [3,11,19,27],[11,19,27,35],[4,12,20,28],
             #/bs diag
             [4,10,16,22],[5,11,17,23],[11,17,23,29],
             [6,12,18,24],[12,18,24,30],[18,24,30,36],
             [7,13,19,25],[13,19,25,31],[19,25,31,37],
             [14,20,26,32],[20,26,32,38],[21,27,33,39]]
    gameOver = False
    mtl = ["X","0"]
    catchWinner = ""
    for i in range(len(mtl)):
        for l in range(len(lanes)):
            if board[lanes[l][0]] == mtl[i] and board[lanes[l][1]] == mtl[i] and board[lanes[l][2]] == mtl[i] and board[lanes[l][3]] == mtl[i]:
                gameOver = True
                catchWinner = mtl[i]
    if gameOver == True:
        announceWinner(catchWinner)
        
    return gameOver
            
def announceWinner(x):
    if x == "X":
        print("You win!")
    if x == "0":
        print("The computer wins!")
    
def checkBoardFull():
    return "_" not in board

boardFull = False

while gameOver == False and boardFull == False:
    gameOver = checkIfGameOver()
    boardFull = checkBoardFull()
    if goFirst == True and gameOver == False and boardFull == False:
        #printBoard()
        playerMove()
        printBoard()
        goFirst = False
    gameOver = checkIfGameOver()
    boardFull = checkBoardFull()
    if goFirst == False and gameOver == False and boardFull == False:
        #printBoard()
        #computerMove() #this is actually a random move but now it's been replaced with strongMove()
        strongMove()
        printBoard()
        goFirst = True
if gameOver == False and boardFull == True:
    print("It's a tie!")
```

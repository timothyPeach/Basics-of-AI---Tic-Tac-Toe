```python
import random as rn
import time

board = ["hey get out of here","_","_","_","_","_","_","_","_","_","what are you doing over here!?"]

#print(row1)
#print(row2)
#print(row3)

info0 = "How to play!!!"
info1 = [1,2,3]
info2 = [4,5,6]
info3 = [7,8,9]
info4= "When it's your turn and you're prompted for a move, use the number guide above!"

infoList = [info0,info1,info2,info3,info4]

for i in infoList:
    print(i)

print("===")
print("===")
print("===")
    
time.sleep(1)

print("Flipping coin to see who goes first!")

gameOver = False

goFirst = True
coin = rn.randint(0,1)
if coin == 0:
    goFirst = False
    
if goFirst == True:
    print("You go first!")
else:
    print("The computer goes first!")
    
def printBoard():
    print("===")
    print("["+board[1]+","+board[2]+","+board[3]+"]")
    print("["+board[4]+","+board[5]+","+board[6]+"]")
    print("["+board[7]+","+board[8]+","+board[9]+"]")
    print("===")
    
def playerMove():
    try:
        move = int(raw_input("Enter an integer 1 thru 9 to determine your move."))
        if board[move] == "_":
            board[move] = "X"
        elif board[move] not in range(1,10):
            print("Hey stick to numbers on the board, lol!")
            playerMove()
        else:
            print("There's already a move there!")
            playerMove()
    except:
        print("You're trying to enter in something crazy, stop dat lol!")
        playerMove()

def computerMove():
    legalMoves = []
    for i in range(len(board[1:10])):
        if board[i] == "_":
            legalMoves.append(i)
    move = rn.choice(legalMoves)
    if board[move] == "_":
        board[move] = "0"
    elif board[move] not in range(1,10):
        print("The computer is trying to make an illegal move! Weird!")
        computerMove()
    else:
        print("The computer is trying to go where a move already exists! Lol!")
        computerMove()
        
def scoreBoard(aBoard):
    lanes = [[1,2,3],[4,5,6],[7,8,9],[1,4,7],[2,5,8],[3,6,9],[1,5,9],[3,5,7]]
    laneStatus = ["","","","","","","",""]

    for i in range(len(lanes)):
        xCount = 0
        zeroCount = 0
        for j in lanes[i]:
            if aBoard[j] == "X":
                xCount += 1
            if aBoard[j] == "0":
                zeroCount += 1
        if xCount == 2 and zeroCount == 0:
            laneStatus[i] = "Danger"
        if xCount == 0 and zeroCount == 2:
            laneStatus[i] = "Great"
        if xCount == 0 and zeroCount == 3:
            laneStatus[i] = "Victory"
        if xCount == 0 and zeroCount == 0:
            laneStatus[i] = "Empty"
        if xCount == 1 and zeroCount == 0:
            laneStatus[i] = "Theirs"
        if xCount == 0 and zeroCount == 1:
            laneStatus[i] = "Ours"
        if xCount > 0 and zeroCount > 0:
            laneStatus[i] = "Dead"

        #for i in laneStatus:
            #print(i)

        # So this process should yield a score.

    currentScore = 0

    for i in laneStatus:
        if i == "Theirs":
            currentScore -= 1
        if i == "Ours":
            currentScore += 1
        if i == "Danger":
            currentScore -= 10
        if i == "Great":
            currentScore += 10
        if i == "Victory":
            currentScore += 100

    #print("currentScore: " + str(currentScore))
    return currentScore

def strongMove():
    legalMoves = []
    for i in range(len(board)):
        if board[i] == "_":
            legalMoves.append(i)
            
    print(legalMoves)
            
    #First of all, I could use machine learning, right, but the game is just too simple so there's 
    #little point not to have the bot be deterministic.
    
    #So the game is comprised of 8 "winning lanes" we'll call them.  The 3 horizontal lanes, the 3 vertical lanes,
    #and the 2 diagonal lanes. When you make the opening move, you want to do it in the center, because you give
    #youself access to 4 winning lanes, while restricting your opponent's victory possibilities to the 4 remaining lanes.
    
    #win groupings by category:
    #horiz: 123,456,789
    #vert: 147,258,369
    #diag: 159,357
    
    move = 0
    
    lanes = [[1,2,3],[4,5,6],[7,8,9],[1,4,7],[2,5,8],[3,6,9],[1,5,9],[3,5,7]]
    
    currentScore = scoreBoard(board)
        
    tempBoards = [0]*len(legalMoves)
    tempScores = [0]*len(legalMoves)    
    
    for i in range(len(tempBoards)):
        tempBoards[i] = []
            
    #print(tempBoards)
    
    for i in range(len(tempBoards)):
        tempBoards[i] = [0]*(len(board)-1)
       
    for x in range(len(tempBoards)):
        for i in range(1,10):
            tempBoards[x][i] = board[i]
            
    for i in range(len(tempBoards)):
        tempBoards[i][legalMoves[i]] = "0"
    
    #print(tempBoards)
    
    for i in range(len(legalMoves)):
        tempScores[i] = scoreBoard(tempBoards[i])
        
    for i in range(len(tempScores)):
        print("move: " + str(legalMoves[i]) + "   score: " + str(tempScores[i]))
        
    move = legalMoves[tempScores.index(max(tempScores))]
    
    #and now we move like before:
    
    if board[move] == "_":
        board[move] = "0"
    elif board[move] not in range(1,10):
        print("The computer is trying to make an illegal move! Weird!")
        #but this time if it screws up, we want it to at least be able to move if it messes up.
        computerMove()
    else:
        print("The computer is trying to go where a move already exists! Lol!")
        #that way the player won't have to sit there, idk.
        computerMove()
        
def checkIfGameOver():
    #win groupings by category:
    #horiz: 123,456,789
    #vert: 147,258,369
    #diag: 159,357
    gameOver = False
    mtl = ["X","0"]
    catchWinner = ""
    for i in range(len(mtl)):
        #this section is horrible and should have been condensed, but it works.
        #horizontals
        if board[1] == mtl[i] and board[2] == mtl[i] and board[3] == mtl[i]:
            gameOver = True
            catchWinner = mtl[i]
        if board[4] == mtl[i] and board[5] == mtl[i] and board[6] == mtl[i]:
            gameOver = True
            catchWinner = mtl[i]
        if board[7] == mtl[i] and board[8] == mtl[i] and board[9] == mtl[i]:
            gameOver = True
            catchWinner = mtl[i]
        #verticals
        if board[1] == mtl[i] and board[4] == mtl[i] and board[7] == mtl[i]:
            gameOver = True
            catchWinner = mtl[i]
        if board[2] == mtl[i] and board[5] == mtl[i] and board[8] == mtl[i]:
            gameOver = True
            catchWinner = mtl[i]
        if board[3] == mtl[i] and board[6] == mtl[i] and board[9] == mtl[i]:
            gameOver = True
            catchWinner = mtl[i]
        #diagonals
        if board[1] == mtl[i] and board[5] == mtl[i] and board[9] == mtl[i]:
            gameOver = True
            catchWinner = mtl[i]
        if board[3] == mtl[i] and board[5] == mtl[i] and board[7] == mtl[i]:
            gameOver = True
            catchWinner = mtl[i]
        #now that all those checks happened:
    if gameOver == True:
        announceWinner(catchWinner)
        
    return gameOver
            
def announceWinner(x):
    if x == "X":
        print("You win!")
    if x == "0":
        print("The computer wins!")
        
moveCounter = 0
    
while gameOver == False and moveCounter != 9:
    gameOver = checkIfGameOver()
    if goFirst == True and gameOver == False and moveCounter != 9:
        #printBoard()
        playerMove()
        printBoard()
        moveCounter += 1
        goFirst = False
    gameOver = checkIfGameOver()
    if goFirst == False and gameOver == False and moveCounter != 9:
        #printBoard()
        #computerMove() #this is actually a random move but now it's been replaced with strongMove()
        strongMove()
        printBoard()
        moveCounter += 1
        goFirst = True
if moveCounter == 9 and gameOver == False:
    print("It's a tie!")
```
